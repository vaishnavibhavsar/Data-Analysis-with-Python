## Data Preparation and Cleaning

Now that we have downloaded the dataset we will do some data preparation and cleaning part.We will remove some columns which we think are unnecessary and one or 2 rows.

So first we will load the dataset into a data frame using Pandas and perform some cleaning operations like exploring number of rows and columns,handle missing and invalid data and many more.



```python
import pandas as pd
import numpy as np
```

```python
claims_df=pd.read_csv('insurance_claims.csv')
```

```python
claims_df
```

```python
claims_df.shape
```

Now the dataset is loaded into a dataframe.It consists of 1000 rows and 40 columns.We did it by using the shape method.The shape method is used to give the number of rows and columns.

```python
claims_df.drop(columns=['months_as_customer','policy_state','policy_csl','umbrella_limit','policy_deductable','insured_zip','insured_education_level','insured_occupation','insured_relationship','capital-gains','capital-loss','incident_type','police_report_available','property_damage','collision_type','incident_severity','incident_state','incident_location','number_of_vehicles_involved','bodily_injuries','witnesses','auto_model','auto_year','fraud_reported','_c39'],axis=1,inplace=True)
```

Now here we will remove unnecessary columns from the dataframe using the drop() function and passing the names of the columns which you want to remove from the datframe.

```python
claims_df
```

```python
claims_df.shape
```

```python
claims_df.columns
```

```python
claims_df.policy_bind_date
```

```python
claims_df['policy_bind_date'] = pd.to_datetime(claims_df.policy_bind_date)
```

```python
claims_df.policy_bind_date
```

Here we have parsed the date using the pd.to_datetime() function in pandas library.We can see now that the datatype of policy_bind_date is datetime64[ns].Previously without parsing it was of object datatype.Parsing date has many advantages.We will use it further as required.

```python
claims_df.insert(loc=0, column='Id', value=np.arange(len(claims_df)))
```

```python
claims_df
```

Now here I have added one more column named 'Id' indicating that the details in that particular row belong to that particular person w.r.t 'Id'.
For ex: If Id=5 it means that the corresponding details in that row are of the 5th person.

```python
claims_df.replace("?",np.NaN)
```

We replaced the missing values by np.NaN if any present in the dataset.

```python
claims_df.isna()
```

Here we used isna() function to find whether there are any missing values in the dataframe.It shows the dataframe with boolean values of True and False.True indicates that there is a missing value and False indicates that there is no missing value.

```python
claims_df.isna().sum()
```

So since the sum of missing values in each column is 0 , it means that there is no missing value

```python
claims_df
```

```python
claims_df=claims_df[['Id','policy_number','age','policy_bind_date','policy_annual_premium','insured_sex','insured_hobbies','incident_date','authorities_contacted','incident_city','incident_hour_of_the_day','total_claim_amount','injury_claim','property_claim','vehicle_claim','auto_make']]
```

## Exploratory Analysis and Visualization

In this section we will have a tour to different types of plots.We will be using Matplotlib ans Seaborn libraries.



Let's begin by importing`matplotlib.pyplot` and `seaborn`.

```python
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

sns.set_style('darkgrid')
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (9, 5)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
```

1.Bar Plot:
We will be plotting bar plot between number of persons in different range of ages having insurance claims

```python
new_claims_df
```

```python
age_df=new_claims_df.groupby('age')[['Id']].count()
age_df
```

So here we have grouped the Id values w.r.t to age and we have used the count() function to count the total number of persons in different age groups. 

```python
age_df=age_df.sort_values('Id').reset_index()
age_df.rename(columns={'Id':'Total'},inplace=True)
```

```python
plt.figure(figsize=(12, 6))
plt.bar(age_df.age,age_df.Total)
plt.xlabel('Age')
plt.ylabel('No of persons having insurance claims')
plt.title('Bar Plot of no of persons having insurance claims with different ages')
```

2. Line Plot: Here will plot the line graph of number of people having different hobbies.But since there are many different hobbies we will consider only 10 values.We will use the head() function.

```python
new_claims_df
```

```python
hobbies_df=new_claims_df.groupby('insured_hobbies')[['Id']].count().head(10)
hobbies_df
```

```python
hobbies_df=hobbies_df.sort_values('Id').reset_index()
hobbies_df.rename(columns={'Id':'Total'},inplace=True)
```

```python
plt.figure(figsize=(18,8))
plt.plot(hobbies_df.insured_hobbies,hobbies_df.Total,'o-b')
plt.xlabel('Hobbies of the insured persons')
plt.ylabel('Total no persons having particular hobbies')
plt.title('Line graph of insured_hobbies vs no of people having different hobbies for 10 people')
```

3. Scatterplot: We want to plot a graph of vehicle claim for various auto make companies.Since there are 1000 rows we cannot plot a line graph.So here we will plot a scatterplot of vehicle claim for various auto make companies.

```python
new_claims_df
```

```python
plt.figure(figsize=(16, 6))
sns.scatterplot(new_claims_df.auto_make,new_claims_df.vehicle_claim,hue=new_claims_df.auto_make,s=100)
plt.title('Vehicle claim for various auto make companies')
```

4.Histogram: We have plotted a histogram to see the distribution of incident hour of the day.

```python
new_claims_df
```

```python
new_claims_df.incident_hour_of_the_day.nunique()
```

```python
new_claims_df.incident_hour_of_the_day
```

```python
plt.hist(new_claims_df.incident_hour_of_the_day)
plt.xlabel('Time in IST')
plt.ylabel('No of incidents')
plt.title('Distribution of incident hour of the day')
```

So here we can say that most of incidents take place after 8pm and in the midnight too.Also there are many incidents in the range of 3pm onwards.

5. BarPlot: We will plot a bar graph between number of insured persons contacting authorities.

```python
new_claims_df
```

```python
authorities_df=new_claims_df.groupby('authorities_contacted')[['Id']].count()
```

```python
authorities_df=authorities_df.sort_values('Id').reset_index()
authorities_df.rename(columns={'Id':'Total'},inplace=True)
```

```python
authorities_df
```

```python
sns.barplot('authorities_contacted','Total',data=authorities_df)
plt.title('Bar plot of no of insured persons contacting authorities')
```

Here we can analyse from the bar plot that above 250 people contacted Police i.e the highest number followed by Fire around more than 200 people contacting them.Then close to 200 people contacting ambulance and other authorities.And near about 100 people contacted none of the authorities.

## Asking and Answering Questions

So here in this section we will ask some interesting questions related to the dataset and try to answer out them using pandas functions and will also answer some of the questions by visualization i.e by plotting graphs.



### Q1: Find out the total number of incidents in different cities.Use a plot of your choice to express the answer.

So here will plot a graph of total number of incidents taking place in different cities.

```python
new_claims_df
```

```python
city_df=new_claims_df.groupby('incident_city')[['Id']].count()
```

We will group the incident city w.r.t Id using the groupby() function  and then count the total number of incidents using the count() method.

```python
new_claims_df.incident_city.nunique()
```

The nunique() method is used to count the number of unique values in the column.So in the incident_city column there are 7 unique cities.

```python
city_df
```

```python
city_df=city_df.sort_values('Id').reset_index()
city_df.rename(columns={'Id':'Total'},inplace=True)
```

So here we sort the values by sort_values() function and we rename the Id column by Total using rename() function. 

```python
city_df
```

```python
plt.plot(city_df.incident_city,city_df.Total,'o-r')
plt.title('Total No of incidents taking place in various cities')
plt.xlabel('Incident City')
plt.ylabel('No of incidents')
```

#### Q2: Calculate the  sum of columns like total claim amount,injury claim amount,property claim amount,vehicle claim amount for each month w.r.t  the policy bind date.

So here we will add a month column and year in the dataset and calculate the sum of  total claim amount, injury claim amount, property claim amount, vehicle claim amount for all the months w.r.t. policy bind date. 

```python
new_claims_df
```

```python
new_claims_df['month']=pd.DatetimeIndex(new_claims_df.policy_bind_date).month
new_claims_df['year']=pd.DatetimeIndex(new_claims_df.policy_bind_date).year
```

So here we will add the two columns month and year using DatetimeIndex() function of pandas library

```python
new_claims_df=new_claims_df[['Id','policy_number','age','policy_bind_date','month','year','policy_annual_premium','insured_sex','insured_hobbies','incident_date','authorities_contacted','incident_city','incident_hour_of_the_day','total_claim_amount','injury_claim','property_claim','vehicle_claim','auto_make']]
```

```python
new_claims_df
```

```python
claims_cost_df=new_claims_df.groupby('month')[['total_claim_amount','injury_claim','property_claim','vehicle_claim']].sum()
```

```python
claims_cost_df
```

```python
claims_cost_df
```

So as we can see that lets say for ex the column of 'total_claim_amount' is having 12 values as of 12 months of policy_bind_date i.e from January to December because we have grouped them w.r.t month. Now the value in the first row is the sum of all the values of 'total_claim_amount' in january.Similarly it goes on for injury_claim, property_claim andvehicle_claim.

#### Q3: Count the number of males and females having insurance claims

So here to answer this question we will plot a pie chart.

```python
new_claims_df
```

```python
gender_df=new_claims_df.groupby('insured_sex')[['Id']].count()
```

```python
gender_df
```

So here we have again grouped the 'insured_sex' column by the 'Id' column by the groupby() function and counted the total number of males and females having insurance_claim

```python
gender_df=gender_df.sort_values('Id').reset_index()
gender_df.rename(columns={'Id':'Total','insured_sex':'gender'},inplace=True)
```

Again here we sorted the values with the sort_values and also reset the index.And also I have renamed the name of the columns.

```python
plt.pie(gender_df.Total,labels=gender_df.gender,autopct='%.2f',startangle=180)
plt.title('Percentage of males and females having insurance claims')
```

So we can see now that females are having more insurance claims than males.The percentage of females is 53.75% of the total persons and percentage of males having insurance claims is 46.25%. We can find out this percentage by the "autopct" parameter.

#### Q4: Calculate the mean of policy annual premium in last 10 years w.r.t policy bind date

```python
new_claims_df
```

```python
premium_mean_df=new_claims_df.groupby('year')[['policy_annual_premium']].mean().tail(10)
```

```python
premium_mean_df
```

So here we have calculated the mean of the column 'policy_annual_premium' of the recent ten years from the dataset w.r.t the policy_bind_date.So we have grouped the column of 'policy_annual_premium' by 'year' column and then calculated the mean of it using the mean() function. We used the tail function and passed 10 so that we get the recent or last 10 years from the year column.

#### Q5: Show with a bar plot the total number of people having vehicles of different manufacturing companies.

```python
new_claims_df
```

```python
new_claims_df.auto_make.unique()
```

unique() function is used to find out the unique values in a particular column.

```python
new_claims_df.auto_make.nunique()
```

So there are 14 unique values i.e 14 unique auto_make companies

```python
vehicle_type_df=new_claims_df.groupby('auto_make')[['Id']].count()
```

```python
vehicle_type_df=vehicle_type_df.sort_values('Id').reset_index()
vehicle_type_df.rename(columns={'Id':'Total','auto_make':'Manufacturing_company'},inplace=True)
```

```python
plt.figure(figsize=(18,8))
plt.title('Bar plot of no of persons having vehicles of different companies')
sns.barplot('Manufacturing_company','Total',data=vehicle_type_df)
```

So here we did the same process of grouping and counting as in previous examples.Now here we plotted the bar graph of Number of people having vehicles of different auto make companies using the sns.barplot() function.
